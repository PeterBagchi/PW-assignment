### Answers to Assignment Questions

1. **What is a parameter?**

Parameters are internal configuration variables whose values are estimated from the data during training (e.g., weights in neural networks or coefficients in linear regression).

2. **What is correlation? What does negative correlation mean?**

Correlation is a statistical measure of the strength and direction of the relationship between two variables. **Negative correlation** means that as one variable increases, the other decreases (an inverse relationship).

3. **Define Machine Learning. What are the main components?**

ML is a branch of AI that enables systems to learn from data and improve from experience without being explicitly programmed. Main components include: **Data**, **Features**, **Model/Algorithm**, and **Evaluation Metrics**.

4. **How does loss value help?**

The loss value measures how far off the model's predictions are from the actual values. A decreasing loss during training indicates the model is learning; a low loss on unseen data suggests the model is 'good' or accurate.

5. **What are continuous and categorical variables?**

- **Continuous:** Numeric values that can take any value within a range (e.g., temperature, price).
- **Categorical:** Variables representing discrete groups or labels (e.g., color, city, gender).

6. **How do we handle categorical variables in Machine Learning? What are the common t
echniques?**

Common techniques include **One-Hot Encoding** (creating dummy binary columns) and **Label/Ordinal Encoding** (converting labels into numerical integers).

7. **What do you mean by training and testing a dataset?**

- **Training:** Using a subset of data to fit the model parameters.
- **Testing:** Using a separate, unseen subset to evaluate how well the model generalizes.

8. **What is sklearn.preprocessing?**

A Scikit-Learn module that provides utility functions to transform raw data into a format suitable for machine learning (e.g., scaling, normalization, encoding).

9. **What is a Test set?**

A portion of the dataset used only at the very end to provide an unbiased evaluation of the final model's performance.

10. **How do we split data?**

In Python, we use `train_test_split` from `sklearn.model_selection`.
**Approach:** Define objective -> Data Collection -> EDA -> Preprocessing -> Model Selection -> Training -> Evaluation -> Deployment.

11. **Why perform EDA?**

Exploratory Data Analysis helps understand data distributions, detect outliers, identify missing values, and find correlations that inform feature selection.

12. **What is correlation?** (Repeated)
 Correlation is a statistical measure of the strength and direction of the relationship between two variables.

13. **What does negative correlation mean?**
**Negative correlation** means that as one variable increases, the other decreases (an inverse relationship).

14. **How to find correlation in Python?**

Using the Pandas library: `df.corr()` calculates the correlation matrix for a DataFrame.

15. **What is causation? Explain difference between correlation and causation with an example.**

Correlation means two things happen together; causation means one *causes* the other. *Example:* Ice cream sales correlate with sunburns (both happen in summer), but ice cream doesn't cause sunburns.

16. **What is an Optimizer?**

An algorithm that adjusts model parameters to minimize the loss function.
- **SGD:** Updates weights for each sample.
- **Adam:** Uses adaptive learning rates for different parameters.

17. **What is sklearn.linear_model?**

A module in Scikit-Learn containing various linear algorithms like Linear Regression, Logistic Regression, and Lasso.

18. **What does model.fit() do?**

It starts the training process. It requires the training features (`X`) and the target labels (`y`) as arguments.

19. **What does model.predict() do? What arguments must be given?**

It uses the trained model to generate predictions on new data. The primary argument is the set of input features (`X`) for which you want predictions.

20. **What are continuous and categorical variables?**
- **Continuous:** Quantitative data that can take any value in a range (e.g., age, weight).
- **Categorical:** Qualitative data that represents specific groups or labels (e.g., fruit type, subscription plan).

21. **What is feature scaling? How does it help?**

Feature scaling is the process of normalizing the range of independent variables. It helps algorithms converge faster (like Gradient Descent) and prevents features with larger magnitudes from dominating those with smaller magnitudes.

22. **How do we perform scaling in Python?**

We typically use `StandardScaler` (for Z-score normalization) or `MinMaxScaler` (for scaling to a range like 0-1) from the `sklearn.preprocessing` library.

23. **What is sklearn.preprocessing?**

It is a module in Scikit-Learn containing functions to change raw feature vectors into a representation that is more suitable for the downstream estimators (scaling, binarization, etc.).

24. **How do we split data for model fitting?**

Using `sklearn.model_selection.train_test_split(X, y, test_size=0.2)`, which randomly divides the dataset into training and testing portions.

25. **Explain data encoding?**

Data encoding is the process of converting categorical data (text labels) into a numerical format so that machine learning algorithms, which require math operations, can process them. Examples include One-Hot Encoding and Label Encoding.
